```python
#merging,joining and concatenating

courses=pd.read_csv('/content/courses.csv')
students=pd.read_csv('/content/students.csv')
nov=pd.read_csv('/content/reg-month1.csv')
dec=pd.read_csv('/content/reg-month2.csv')
delivery=pd.read_csv('/content/deliveries.csv')
matches=pd.read_csv('/content/matches.csv')
```

```python
#pd.concat([nov,dec])# it scales horizontally
regs=pd.concat([nov,dec],ignore_index=True)
#nov.appned(dec,ignore_index=True)# both are same , use concat for better practice
regs.tail(10)
```

```python
pd.concat([nov,dec],axis=1)# it scales vertically

```

```python
multi=pd.concat([nov,dec],keys=['Nov','Dec'])
#multindex DataFrame
multi.loc['Nov']
multi.loc[('Nov',0)]
multi.loc[('Dec',9)]
multi.loc[('Nov',0):('Dec',9)]
```

```python
pd.concat([students,regs],axis=1)
```

```python
#merge
students.merge(regs, on='student_id',how='inner').head(5)
```

```python
courses.merge(regs,on='course_id',how='left').head(5)
```

```python
courses.merge(regs,on='course_id',how='right').head(5)
```

```python
#find revenue
regs.merge(courses,on='course_id',how='inner')['price'].sum()
```

```python
# find month by month revenue
temp_df=pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
temp_df.merge(courses,on='course_id',how='inner').groupby('level_0')['price'].sum()
```

```python
regs.head(5)
```

```python
students.merge(regs,on='student_id').merge(courses,on='course_id')[['name','course_name','price']].groupby('name')['price'].sum()
```

```python
regs.merge(courses,on='course_id',how='inner').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
s_id=np.intersect1d(dec['student_id'],nov['student_id'])
students=students[students['student_id'].isin(s_id)]
students[['student_id','name']]

# array([ 1,  3,  7, 11, 16, 17, 18, 22, 23])
```

```python
c_id=np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(c_id)]['course_name']
#courses that are not enrolled
```

```python
regs.merge(students,on='student_id').groupby(['student_id','name'])['name'].count().sort_values(ascending=False).head(3)
# top 3 students with most enrollment
```
