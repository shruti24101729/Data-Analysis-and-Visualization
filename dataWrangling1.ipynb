```python
##### pandas export-to csv,json,html,json,sql,excel

import pandas as pd
df=pd.DataFrame()
```

```python
#csv
df=pd.read_csv('/content/deliveries.csv')
df.info()
```

```python
temp_df=df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('top_run.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('batsman_vs_team.csv')
```

```python
#excel
temp_df.to_excel('top_run.xlsx',index=False)
```

```python
with pd.ExcelWriter('top_run_and_batsman.xlsx') as writer:
  temp_df.to_excel(writer,sheet_name='top_run',index=False)
```

```python
# html

df.pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count')
temp=df.query('batsman_runs==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count')
temp.to_html('batsman_vs_team.html')
```

```python
#json
temp_df.to_json('top_run.json')
```
